```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_monthly_mean_price_by_year(df):
    """
    Plots the mean price range per month, separated by year, using line plots.

    Args:
        df (pd.DataFrame): The input DataFrame with 'listing_first_date' and 'price' columns.
    """
    df = df.copy()

    # Convert listing_first_date to datetime
    df['listing_first_date'] = pd.to_datetime(df['listing_first_date'])

    # Extract year and month
    df['year'] = df['listing_first_date'].dt.year
    df['month'] = df['listing_first_date'].dt.month

    # Group by year and month, then calculate the mean price
    monthly_mean_price = df.groupby(['year', 'month'])['price'].mean()

    # Get unique years
    years = df['year'].unique()

    # Create a figure and a set of subplots, one for each year
    fig, axes = plt.subplots(nrows=1, ncols=len(years), figsize=(15, 5), sharey=True)


    for ax, year in zip(axes, years):

        # Filter data for the current year
        year_data = monthly_mean_price[year]

        # Plot the data for the current year
        ax.plot(year_data.index, year_data.values, marker='o', linestyle='--',color = 'purple')

        ax.set_title(f'Mean Price in {year} for Active New Vehicles')
        ax.set_xlabel('Month')
        ax.set_ylabel('Mean Price')
        ax.set_xticks(range(1, 13))
        ax.set_xticklabels(
            ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
            rotation=45
        )
        ax.grid(True)

    plt.tight_layout()
    plt.show()


# Example usage with your df_active_new DataFrame:
plot_monthly_mean_price_by_year(df_active_new)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Assuming your DataFrame is named df_active
def plot_avg_price_by_make(df):
    """
    Plots the average price by car make using a bar graph.

    Args:
        df (pd.DataFrame): The input DataFrame with 'make' and 'price' columns.
    """
    # Calculate the average price for each car make
    avg_price_by_make = df.groupby('make')['price'].mean().sort_values(ascending=False)

    # Create the bar plot
    plt.figure(figsize=(12, 6))
    avg_price_by_make.plot(kind='bar', color='green')
    plt.title('Average Price by Active New Car Make')
    plt.xlabel('Car Make')
    plt.ylabel('Average Price')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Example usage with your df_active DataFrame:
plot_avg_price_by_make(df_active_new)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_monthly_mean_price_by_year(df):
    """
    Plots the mean price range per month, separated by year, using line plots.

    Args:
        df (pd.DataFrame): The input DataFrame with 'listing_first_date' and 'price' columns.
    """
    df = df.copy()

    # Convert listing_first_date to datetime
    df['listing_first_date'] = pd.to_datetime(df['listing_first_date'])

    # Extract year and month
    df['year'] = df['listing_first_date'].dt.year
    df['month'] = df['listing_first_date'].dt.month

    # Group by year and month, then calculate the mean price
    monthly_mean_price = df.groupby(['year', 'month'])['price'].mean()

    # Get unique years
    years = df['year'].unique()

    # Create a figure and a set of subplots, one for each year
    fig, axes = plt.subplots(nrows=1, ncols=len(years), figsize=(15, 5), sharey=True)


    for ax, year in zip(axes, years):

        # Filter data for the current year
        year_data = monthly_mean_price[year]

        # Plot the data for the current year
        ax.plot(year_data.index, year_data.values, marker='o', linestyle='--',color = 'purple')

        ax.set_title(f'Mean Price in {year} for Active New Ford F150')
        ax.set_xlabel('Month')
        ax.set_ylabel('Mean Price')
        ax.set_xticks(range(1, 13))
        ax.set_xticklabels(
            ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
            rotation=45
        )
        ax.grid(True)

    plt.tight_layout()
    plt.show()


# Example usage with your df_active_new DataFrame:
plot_monthly_mean_price_by_year(df_active_new[(df_active_new['make'] == 'Ford') & (df_active_new['model'] == 'F150')])
```
